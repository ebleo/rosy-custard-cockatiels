```python
%matplotlib inline
import pandas as pd
import numpy as np
import sklearn
```

```python
cols = ['ID', 'clump thick.', 'unif.ty cell size', 'unif.ty cell shape', 'marg. adhesion', \
        'single epithelial cell sz', 'bare nuclei', 'bland chromatin', 'norm. nucleoli',\
       'mitoses', 'class']
```

```python

df = pd.read_csv('../ipynb/Data/breast-cancer-wisconsin.data.csv')
df.columns = cols
```

```python
df.describe()
```

```python
for col in df:
    print('{:<30}'.format(col), '\n', df[col].value_counts())
```

```python
t_df = df
```

```python
t_df = t_df[t_df['bare nuclei'] != '?'] # dropping unknown values
```

```python
t_df['bare nuclei'].value_counts(sort=True)
```

```python
t_df.describe()
# removal of '?' only affected 'bare nuclei' (i think)
```

```python
t_df.groupby('class').size()
```

```python
from sklearn import svm
lin_clf = svm.LinearSVC()
```

need to sep target from data


```python
train = t_df.iloc[:, :-1]
train.describe()
```

```python
# target = pd.DataFrame(data=t_df.loc[:,'class'])
target = t_df.loc[:,'class']
target.describe()
```

```python
for col in train.iloc[:,1:].columns: # omitting ID col
    print(train[col].value_counts(dropna=False))
```

```python
X, y = train, target
print(repr(lin_clf.fit(X[:-10], y[:-10])))
```

```python
lin_clf.predict(X[-5:])
```

```python
plist = lin_clf.predict(X[-5:])
for item in plist:
    print(item)

```

```python
zippy = zip(plist, y[-5:])
```

```python
# print(y[-5:])
zippy = list(zip(plist, y[-5:]))
for p,yval in zippy:
    print('Predict: ',p,'  Actual:  ',yval)
# does not appear accurate, need to improve model
```

```python
# for saving model
import pickle
s = pickle.dumps(lin_clf)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lin_clf, X, y, scoring='accuracy')
```
